```python
def Predict_Aspect(tf,data,data_uji):
    text_tf=tf.fit_transform(data['Review'].astype('U'))
    from sklearn.model_selection import train_test_split
    X_train,X_test,y_train,y_test=train_test_split(text_tf,data['Aspek'],test_size=0.2,random_state=42)
    from sklearn.naive_bayes import MultinomialNB
    clf = MultinomialNB().fit(X_train,y_train)
    text_hm=tf.transform(data_uji['Review'].astype('U'))
    heh=clf.predict(text_hm)
    print(heh)

def Predict_Sentiment(tf,data,data_uji):
    text_tf=tf.fit_transform(data['Review'].astype('U'))
    from sklearn.model_selection import train_test_split
    X_train,X_test,y_train,y_test=train_test_split(text_tf,data['Sentiment'],test_size=0.2,random_state=42)
    from sklearn.naive_bayes import MultinomialNB
    clf = MultinomialNB().fit(X_train,y_train)
    text_hm=tf.transform(data_uji['Review'].astype('U'))
    heh=clf.predict(text_hm)
    print(heh)
    
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
tf = TfidfVectorizer()
def combine(tf):
    data=pd.read_csv('data_sentiment.csv',encoding='latin1')
    data_uji=pd.read_csv(r"C:\Users\win10\Documents\review.csv")
    sentimen=Predict_Sentiment(tf,data,data_uji)
    aspek=Predict_Aspect(tf,data,data_uji)
    return{'sentimen':sentimen,'aspek':aspek}

print(combine(tf))
    
```
